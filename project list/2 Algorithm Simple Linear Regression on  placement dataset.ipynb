```python
# Splitting variables
X = place_df.iloc[:,:1]
y = place_df.iloc[:,1:]

# Here X: Features (input values)

# and y: Labels or target (output values to predict)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression # Regressor model import linear regression liberry 

#70% of the data → used for training (X_train, y_train)
# 30% of the data → used for testing (X_test, y_test)
#random_state=0 makes sure you get the same split every time (it's like setting a seed)

X_train ,X_test , y_train, y_test = train_test_split(X,y ,test_size=0.3,random_state= 0)
X_train ,X_test , y_train, y_test
```

```python
# Hey model! Learn from this training data.
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
# Model makes predictions on test data
y_pred_X_test = reg.predict(X_test)  # Predict output from X_test
from sklearn.metrics import mean_squared_error
# Then compare predictions to the real values:

print("MSE:", mean_squared_error(y_test, y_pred_X_test))

```

```python
#Step-by-step to compare y_pred and y_test:
# flatten() is used to make sure the values are in a simple 1D format for comparison.


import pandas as pd

comparison = pd.DataFrame({
    'Actual': y_test.values.flatten(),
    'Predicted': y_pred_X_test.flatten()
})

print(comparison)
```

```python

# Regressor coefficients and intercept
print(f'Cofficient:{ reg.coef_}')
print(f'Intercept:{reg.intercept_}')
print('y = ', reg.intercept_, '+', reg.coef_, '* X')
#reg.coef_ → the slope of the line (how much y changes per unit of X)
#reg.intercept_ → where the line crosses the y-axis (value of y when X is 0)

#This formula tells you how your model is making predictions.
#You can plug in any x to calculate the predicted y
#This helps understand the relationship between input and output
```

```python

# evaluate the model
print('MSE = ', mean_squared_error(y_test, y_pred_X_test ))
print('R2 = ', r2_score(y_test, y_pred_X_test ))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred_X_test )))
```

```python
# 🧪 1. MSE – Mean Squared Error
# MSE calculates the average of the squared errors between actual and predicted values.
# Lower MSE means a better model.
# It's in square units, so it's harder to interpret directly.

# Example:
# Real values: 10, 15, 20
# Predicted:   12, 13, 22
# Errors: (12–10)² + (13–15)² + (22–20)² = 4 + 4 + 4 = 12
# MSE = 12 ÷ 3 = 4

# 🧮 2. RMSE – Root Mean Squared Error
# RMSE is the square root of MSE.
# It brings the error back to the original unit of the target (like dollars, kg, etc.)
# Easier to understand: RMSE = 4 means you're off by ~4 units on average.

# 📊 3. R² – R-Squared (Coefficient of Determination)
# R² tells how well the model explains the variation in the data.
# Values range from 0 to 1 (or can be negative if the model is very bad).
# 1.0 = perfect predictions
# 0.8 = very good
# 0.5 = okay
# 0 or below = model is not helpful
```
