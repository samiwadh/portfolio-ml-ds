```python
#labelEncoder from scikit-learn (sklearn) is a useful tool for encoding target labels. It 
# #assigns a unique integer value to each label, making it easier to work with categorical 
# #data in machine learning models. 

                            #What this code does labelEncoder:
#It applies Label Encoding to all categorical (non-numeric) columns in your DataFrame (sea_df), 
#converting text labels (like "male", "female", "S", "C", etc.) into numeric values (like 0, 1, 2, etc.),
#which are required for machine learning models.
```

```python
from sklearn.preprocessing import LabelEncoder  # Import LabelEncoder for converting categorical labels into numbers

# Loop through each column in the DataFrame
for col in sea_df.columns:
    # Check if the column's data type is 'object' (usually text) or a 'category'
    if sea_df[col].dtype == 'object' or sea_df[col].dtype.name == 'category':
        
        # Make sure all values in the column are of allowed types (str, int, float) to avoid encoding errors
        if sea_df[col].apply(lambda x: isinstance(x, (str, int, float))).all():
            
            # Convert the column to string type (to ensure uniformity), then apply Label Encoding
            sea_df[col] = LabelEncoder().fit_transform(sea_df[col].astype(str))


sea_df.head()
```

```python

# X and y column
X = sea_df.drop('survived', axis=1)
y = sea_df['survived'] 

# Here X: Features (input values)

# and y: Labels or target (output values to predict)
```

```python
from sklearn.model_selection import train_test_split


# Split the data: 80% for training, 20% for testing

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# Optional comment for understanding
# 80% of data â†’ training (X_train, y_train)
# 20% of data â†’ testing (X_test, y_test)
```

```python
# ðŸ” Why we do this:
# Machine learning models (like LogisticRegression) require numeric input and can't handle missing or string values.
# So, we convert object types to numeric and fill missing values to ensure smooth training.


# Convert the 'age' column to numeric type in both training and test sets
# If any value can't be converted (like a string or typo), it will be replaced with NaN
X_train['age'] = pd.to_numeric(X_train['age'], errors='coerce')
X_test['age'] = pd.to_numeric(X_test['age'], errors='coerce')

# Fill any NaN values (e.g., from conversion errors or missing values) with the mean age from the training set
# This ensures the model doesn't break due to missing or non-numeric values
X_train['age'] = X_train['age'].fillna(X_train['age'].mean())
X_test['age'] = X_test['age'].fillna(X_train['age'].mean())  # use train mean to avoid data leakage


```

```python
sea_df.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression

log_rec = LogisticRegression().fit(X_train, y_train)
log_rec
```

```python
y_pred_test = log_rec.predict(X_test)
y_pred_test
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix, classification_report

# âœ… Accuracy: How many predictions were correct overall
accuracy = accuracy_score(y_test, y_pred_test)

# âœ… Recall: How many actual positives were correctly identified (Sensitivity)
recall = recall_score(y_test, y_pred_test)

# âœ… Precision: How many predicted positives were actually correct
precision = precision_score(y_test, y_pred_test)

# âœ… F1 Score: Harmonic mean of precision and recall (balanced metric)
f1 = f1_score(y_test, y_pred_test)

# âœ… Confusion Matrix: Table showing TP, FP, FN, TN
conf_matrix = confusion_matrix(y_test, y_pred_test)

# âœ… Classification Report: Includes precision, recall, f1-score, and support for each class
class_report = classification_report(y_test, y_pred_test)

# Print results
print("ðŸ”¹ Accuracy:", accuracy)
print("ðŸ”¹ Recall:", recall)
print("ðŸ”¹ Precision:", precision)
print("ðŸ”¹ F1 Score:", f1)
print("\nðŸ”¹ Confusion Matrix:\n", conf_matrix)
print("\nðŸ”¹ Classification Report:\n", class_report)
```
