# Univariate Analysis

```python
# prices distribtion 

data=airbnb['price']
```

```python
# identying outlieres in price
df = airbnb[airbnb['price']<1600]
sns.boxplot(data=df, x='price')
```

```python
# price distrubtion 
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")  # clean background
sns.histplot(data=df, x='price', color='red', edgecolor='black', alpha=0.7, bins=30)
plt.ylabel("Frequency")


plt.show()
```

```python
df = airbnb[airbnb['reviews_per_month']>4.00]
sns.boxplot(data=df , x ='reviews_per_month')
plt.title("reviews_per_month")
plt.show()
```
